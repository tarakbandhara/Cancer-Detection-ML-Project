```python
# Importing Libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore', message="X does not have valid feature names")
```

```python
data = pd.read_csv('data.csv')
```

```python
# Performing EDA
```

```python
data.info()
```

```python
data.isnull().sem()
```

```python
data.head()
```

```python
data.tail()
```

```python
# Moving target variable to last
cols = list(data.columns.values)
cols.pop(cols.index('diagnosis'))
data = data[cols+['diagnosis']]
```

```python
data.head()
```

```python
# Droping unwanted columns
data = data.drop(['id', 'Unnamed: 32'], axis=1).copy()
data.info()
```

```python
# Data Preprocessing
```

```python
# Adding data to 'x' and 'y'
x = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
# Doing feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# Fit the scaler on the training data and transform both training and test data
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Training the Logistic Regression Model
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train, y_train)
```

```python
# Model Evaluation
```

```python
# Accuracy Score
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Training Data Accuracy
x_train_prediction = regressor.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print(f"Training Data Accuracy : {training_data_accuracy:.2f}")
```

```python
# Test Data Accuracy
x_test_prediction = regressor.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print(f"Test Data Accuracy : {test_data_accuracy:.2f}")
```

```python
# Plotting training and test accuracy
accuracy_scores = [training_data_accuracy, test_data_accuracy]
labels = ['Training Accuracy', 'Test Accuracy']
bars = plt.bar(labels, accuracy_scores, color=['red', 'orange'])
plt.ylabel('Accuracy')
plt.title('Training vs Test Data Accuracy')
plt.ylim(0.9, 1.0) # Set y-axis limits for better visualization of differences

# Add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, f'{yval:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# Analyze model performance based on training and test accuracy
print("Model Performance Analysis :\n")
print(f"Training Data Accuracy: {training_data_accuracy:.2f}\n")
print(f"Test Data Accuracy: {test_data_accuracy:.2f}")


if training_data_accuracy > test_data_accuracy + 0.05:
    print("\nResult : The model might be overfitted. It performs very well on the training data but significantly worse on unseen test data.☹️")
elif test_data_accuracy < training_data_accuracy - 0.05 and test_data_accuracy < 0.8:
     print("\nResult : The model might be underfitted. It does not perform well on either the training or the test data.☹️")
else:
    print("\nResult : The model appears to be properly fitted. The training and test accuracies are close, indicating good generalization.🙂")
```

```python
y_pred = regressor.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print (f"Accuracy Score : {accuracy_score(y_test, y_pred)}")
```
